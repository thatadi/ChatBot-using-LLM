```python
from transformers import pipeline
import pandas as pd

# Reading our dataset into a dataframe
table =pd.read_csv('LLM-Sample-Input-File.csv')
```

```python
table
```

```python
# Since category column has the same values lets drop it
table=table.drop(columns=['Category'])

# Renaming the last column so the model can understand that revenue is to be extracted from this column
table=table.rename(columns={' Value - Randomized ' : 'revenue'})
table
```

```python
question = "How much revenue did Potato Inc. make from selling Tablets in Q1 2022"

tqa = pipeline(task="table-question-answering", model="google/tapas-large-finetuned-wtq")

print(tqa(table=table, query=question)['cells'][0])
# answer should be --> 57,984,000,000
```
